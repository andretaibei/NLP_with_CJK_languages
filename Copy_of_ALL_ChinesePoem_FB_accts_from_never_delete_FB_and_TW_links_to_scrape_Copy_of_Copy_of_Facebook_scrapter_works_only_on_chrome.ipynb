```python
#Time.sleep SOLUTION NEVER DELETE = https://github.com/kevinzg/facebook-scraper/issues/310#issuecomment-852652846
#Package Creator's code to get MAXIMUM Posts while avoiding Facebook Block from = https://github.com/kevinzg/facebook-scraper/issues/625#issuecomment-1006988097
from facebook_scraper import *
import time
start = time.time()
posts = []
while True:
  try:
    for post in get_posts("147251908669637", page_limit=None, timeout=60, options={"allow_extra_requests": False, "posts_per_page": 200}):
      posts.append(post['text'])
    break
  except exceptions.TemporarilyBanned:
    print("Temporarily banned, sleeping for 10m")
    time.sleep(600)
    print(f"{len(posts)} posts retrieved in {round(time.time() - start)}s. Oldest post: {posts[-1].get('time')}")
```
