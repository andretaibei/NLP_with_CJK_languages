```python
#Package Creator's code to get MAXIMUM Posts while avoiding Facebook Block from = https://github.com/kevinzg/facebook-scraper/issues/625#issuecomment-1006988097
import time
start = time.time()
posts = []
try:
  for post in get_posts("325866880804258", pages=200, timeout=60, options={"allow_extra_requests": False, "posts_per_page": 200}):
    posts.append(post['text'])
except:
    print(f"{len(posts)} posts retrieved in {round(time.time() - start)}s. Oldest post: {posts[-1].get('time')}")
```

```python

#Directions and steps from = https://towardsdatascience.com/korean-natural-language-processing-in-python-cc3109cbbed8
!pip install konlpy
import konlpy
from konlpy.tag import Okt
from konlpy.utils import pprint
```

```python
posts_in_string_format = " ".join(str(x) for x in posts)
```

```python
pprint(okt.pos(posts_in_string_format, norm=True, stem=True))
```
